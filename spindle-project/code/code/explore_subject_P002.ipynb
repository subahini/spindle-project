```python
from scipy.io import loadmat

# Load your .mat file
mat = loadmat("../data/raw/SeLECTS_P002_1_epochsScoring_artndxn.mat")
epo = loadmat("../data/raw/SeLECTS_P002_1_epochsScoring .mat")
# Show all top-level keys
print("üì¶ Keys in the .mat file:")
print(mat.keys())
print(epo.keys())
```

```python
artndxn = mat['artndxn'].flatten()

print(f"üîç Total epochs: {len(artndxn)}")
print(f"üü¢ Clean epochs (0): {(artndxn == 0).sum()}")
print(f"üî¥ Artifact epochs (1): {(artndxn == 1).sum()}")

# Plot it
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 1))
plt.plot(artndxn, '|', markersize=10)
plt.title("Artifact Index (1 = bad, 0 = clean)")
plt.xlabel("Epoch index")
plt.yticks([0, 1])
plt.show()
```
